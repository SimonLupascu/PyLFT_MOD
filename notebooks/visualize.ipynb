```python
import py3Dmol
import os
import glob

notebook_dir = os.path.dirname(os.path.abspath('__file__'))

mol2_files = sorted(glob.glob(os.path.join(notebook_dir, '../src/pylft_mod/*.mol2')))

if not mol2_files:
    print("No .mol2 files found. Run architector_builder.py first, then re-run this cell.")
else:
    print("Available complexes:")
    for i, f in enumerate(mol2_files):
        print(f"  [{i}] {os.path.basename(f)}")

    search = "trans"

    matches = [f for f in mol2_files if search in os.path.basename(f)]

    if not matches:
        print(f"No file found containing '{search}'. Run the builder first.")
    else:
        mol2_path = matches[0]
        print(f"Showing: {os.path.basename(mol2_path)}")

        with open(mol2_path) as f:
            xyz_data = f.read()

        viewer = py3Dmol.view(width=700, height=500)
        viewer.addModel(xyz_data, 'mol2')
        viewer.setStyle({'sphere': {'scale': 0.3}, 'stick': {'radius': 0.15}})

        axis_length = 4.0
        axis_radius = 0.05

        viewer.addCylinder({'start': {'x': -axis_length, 'y': 0, 'z': 0},
                            'end':   {'x':  axis_length, 'y': 0, 'z': 0},
                            'radius': axis_radius, 'color': 'red', 'opacity': 0.8})
        viewer.addLabel('X', {'position': {'x': axis_length + 0.3, 'y': 0, 'z': 0},
                              'fontColor': 'red', 'fontSize': 14, 'backgroundColor': 'transparent'})

        viewer.addCylinder({'start': {'x': 0, 'y': -axis_length, 'z': 0},
                            'end':   {'x': 0, 'y':  axis_length, 'z': 0},
                            'radius': axis_radius, 'color': 'green', 'opacity': 0.8})
        viewer.addLabel('Y', {'position': {'x': 0, 'y': axis_length + 0.3, 'z': 0},
                              'fontColor': 'green', 'fontSize': 14, 'backgroundColor': 'transparent'})

        viewer.addCylinder({'start': {'x': 0, 'y': 0, 'z': -axis_length},
                            'end':   {'x': 0, 'y': 0, 'z':  axis_length},
                            'radius': axis_radius, 'color': 'blue', 'opacity': 0.8})
        viewer.addLabel('Z', {'position': {'x': 0, 'y': 0, 'z': axis_length + 0.3},
                              'fontColor': 'blue', 'fontSize': 14, 'backgroundColor': 'transparent'})

        viewer.zoomTo()
        viewer.show()
```
